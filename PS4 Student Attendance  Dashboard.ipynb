### Date and Ranked Columns

finding date columns and ranked columns

data columns are non numerical.if any data is missing in date columns we need replace the empty value with highest no of occurances in that column

ranked columns are numerical. if any data is missing in ranked columns we need to replace with the avg rank 

### Finding Date and Ranked columns for class 1 data

```python
# Date columns for class 1 data
class_1_date_columns = []
def check_date_column(column_name):
    date_string = column_name
    formt = "%d-%m-%y"
    try:
        res = bool(datetime.datetime.strptime(date_string, formt))
    except ValueError:
        res = False
    return res
for i in class_1_data.columns:
    if check_date_column(i):
        class_1_date_columns.append(i)
print(class_1_date_columns)
```

```python
# finding total no of classes in class 1 data
no_of_classes_in_class_1 = len(class_1_date_columns)
no_of_classes_in_class_1
```

```python
# Ranked columns for class 1 data
class_1_ranked_columns = []
def check_ranked_column(column_name):
    first_letter = column_name[0]
    date = column_name[2:]
    if first_letter == "R" and check_date_column(date):
        return True
    else:
        return False
for i in class_1_data.columns:
    if check_ranked_column(i):
        class_1_ranked_columns.append(i)
print(class_1_ranked_columns)
```

```python
# finding total no of ranked columns in class 1 data
no_of_ranked_columns_in_class_1 = len(class_1_ranked_columns)
print(no_of_ranked_columns_in_class_1)
```

### Finding Date and Ranked columns for class 2 data

### Dealing other missing values in class 1 data

```python
# finding columns where data is missing in particular row or rows for class 1 data
class_1_missed_data_columns = []
for i in class_1_data.columns:
    if class_1_data[i].isnull().sum() > 0:
        class_1_missed_data_columns.append(i)
print(class_1_missed_data_columns)
```

```python
for i in class_1_missed_data_columns:
    if i in class_1_date_columns:
        temp_dict = dict(class_1_data[i].value_counts())
        value = max(temp_dict,key = temp_dict.get)
        # replace empty value with highest occurance of that column
        class_1_data[i] = class_1_data[i].fillna(value)
    elif i in class_1_ranked_columns:
        avg_rank_class_1 = (class_1_data[class_1_ranked_columns].max().max() + class_1_data[class_1_ranked_columns].min().min()) // 2
        # replacing empty column with student avg rank
        class_1_data[i] = class_1_data[i].fillna(avg_rank_class_1)
        
```

Now class 1 data doesn't have missing values or empty values

```python
class_1_data.isnull().sum()
```

### Dealing other missing values in class 2 data

### finding each student avg rank and class avg rank in class 1 data

```python
# finding each student avg rank for class_1_data
class_1_student_rank_sum = {}
class_1_student_avg_rank_dict = {}
for i in class_1_student_roll_nums:
    sum_of_ranks = list(class_1_data.loc[class_1_data['Student Roll Num'] == i][class_1_ranked_columns].sum(axis=1))
    class_1_student_rank_sum[i] = sum_of_ranks[0]
    class_1_student_avg_rank_dict[i] = (sum_of_ranks[0]/(no_of_classes_in_class_1 * 10))*100
print(class_1_student_avg_rank_dict)
```

```python
# finding class average among all the students for class 1 data
avg_class_rank_for_class_1 = (sum(class_1_student_rank_sum.values())/(total_no_of_students_in_class_1 * no_of_classes_in_class_1 * 10))*100
print(avg_class_rank_for_class_1)
```

### finding each student avg rank and class avg rank in class 2 data

```python
# finding each student avg rank for class_2_data
class_2_student_rank_sum = {}
class_2_student_avg_rank_dict = {}
for i in class_2_student_roll_nums:
    sum_of_ranks = list(class_2_data.loc[class_2_data['Student Roll Num'] == i][class_2_ranked_columns].sum(axis=1))
    class_2_student_rank_sum[i] = sum_of_ranks[0]
    class_2_student_avg_rank_dict[i] = (sum_of_ranks[0]/(no_of_classes_in_class_2 * 10))*100
print(class_2_student_avg_rank_dict)
```

```python
# finding class average among all the students for class 1 data
avg_class_rank_for_class_2 = (sum(class_2_student_rank_sum.values())/(total_no_of_students_in_class_2 * no_of_classes_in_class_2 * 10))*100
print(avg_class_rank_for_class_2)
```

### finding required dataset for class 1 data

 finding count of missed classed and ccount of attended classes for each student in class 1 data

```python
final_student_data_for_class_1 = {}
for i in class_1_student_roll_nums:
    attendance = dict(class_1_data.loc[class_1_data['Student Roll Num'] == i][class_1_date_columns].sum(axis=0).value_counts())
    final_student_data_for_class_1[i]=attendance
    if 'Missed' not in attendance:
        final_student_data_for_class_1[i]['Missed'] = 0
    if 'Attended' not in attendance:
        final_student_data_for_class_1[i]['Attended'] = 0
for i in class_1_student_roll_nums:
    final_student_data_for_class_1[i]['Student Roll Num'] = i
final_student_data_for_class_1
```

adding each student avg rank in dicionary

```python
for i in final_student_data_for_class_1:
    if i in class_1_student_avg_rank_dict:
        final_student_data_for_class_1[i]['Student Avg Rank'] = class_1_student_avg_rank_dict[i]
        final_student_data_for_class_1[i]['Class Avg Rank'] = avg_class_rank_for_class_1
final_student_data_for_class_1
```

### Performance Measure 

Here i am taking some test cases

if student having more than class avg rank and he attended more than 4 classes he is under Very Satisfied categeory

if student having more than class avg rank and he attended less than 4 classes he is under  Satisfied categeory

if student having lesss than class avg rank and he attended less than 4 classes he is under Un Satisfied categeory

if student having more than class avg rank and he attended less than 4 classes he is under Satisfied categeory

### finding performance of each student in class 1 data

```python
for i in final_student_data_for_class_1:
    if final_student_data_for_class_1[i]['Student Avg Rank'] < avg_class_rank_for_class_1 and final_student_data_for_class_1[i]['Attended'] <=4:
        final_student_data_for_class_1[i]["Performance"] = 'Un Satisfied'
    elif final_student_data_for_class_1[i]['Student Avg Rank'] >= avg_class_rank_for_class_1 and final_student_data_for_class_1[i]['Attended'] <=4:
        final_student_data_for_class_1[i]["Performance"] = 'Satisfied'
    elif final_student_data_for_class_1[i]['Student Avg Rank'] >= avg_class_rank_for_class_1 and final_student_data_for_class_1[i]['Attended'] > 4:
        final_student_data_for_class_1[i]["Performance"] = 'Very Satisfied'
    elif final_student_data_for_class_1[i]['Student Avg Rank'] < avg_class_rank_for_class_1 and final_student_data_for_class_1[i]['Attended'] > 4:
        final_student_data_for_class_1[i]["Performance"] = 'Un Satisfied'
final_student_data_for_class_1
```
